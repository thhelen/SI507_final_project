```python
playlist_link = "https://open.spotify.com/playlist/3oDWJvExLZlWOseBy3jvd7?si=40cd2bd2070a4104"
playlist_URI = playlist_link.split("/")[-1].split("?")[0]

track_uris = []
track_name, artist_name, album, track_pop, track_year=[],[],[],[],[]
offset = 0
limit = 100

while True:
    response = sp.playlist_tracks(playlist_URI, offset=offset, limit=limit)
    tracks = response["items"]

    for track in tracks:
        track_name.append(track["track"]["name"])
        
        artist_name.append(track["track"]["artists"][0]["name"])
    
        album.append(track["track"]["album"]["name"])
        
        track_pop.append(track["track"]["popularity"])

        track_year.append(track["track"]["album"]["release_date"])
    
        track_uris.append(track["track"]["uri"])
        
    if len(tracks) < limit:
        break
        
    offset += limit
```

```python
final_list=[]
```

```python
# changing the range of years multiple times to get all years possible
for year in range(1998,2021):
    print(year)
    response = requests.get('http://en.wikipedia.org/wiki/Billboard_Year-End_Hot_100_singles_of_'+str(year))
    page = response.text
    soup = BeautifulSoup(page, 'html.parser')
    tables_wikitable = soup.find_all('table', 'wikitable')
    rows = [row for row in tables_wikitable[0].find_all('tr')]
    def get_td(row):
        return [td for td in row.find_all('td')]

    def rank(td):
        return td[0].string

    def artist_name(td):
        return td[2].a.string

    def title(td):
        if td[1].a:
            return td[1].a['title']
        else:
            return td[1].string
        
    td_list = [get_td(row)for row in rows[1:]]
    list_of_dicts = [{'rank':rank(td), 'artists': artist_name(td), 'title':title(td), 'year': year} for td in td_list]
    final_list.append(list_of_dicts)
```

```python
final_list_flat = [item for sublist in final_list for item in sublist]
print(final_list_flat)
```
